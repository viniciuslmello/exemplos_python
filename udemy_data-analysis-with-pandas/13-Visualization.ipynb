### Creating Bar Graphs to Show Counts

```python
import yfinance as yf
yf.pdr_override()
# y_symbols = ['SCHAND.NS', 'TATAPOWER.NS', 'ITC.NS']
y_symbols = ['BB']
from datetime import datetime
startdate = datetime(2007,1,1)
enddate = datetime(2020,12,31)
bb = pdr.get_data_yahoo(y_symbols, start=startdate, end=enddate)
bb
```

```python
def rank_performance(stock_price):
    if stock_price <= 10:
        return "Poor"
    elif stock_price <= 50:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
bb["Close"].apply(rank_performance)
```

```python
bb["Close"].apply(rank_performance).value_counts()
```

```python
bb["Close"].apply(rank_performance).value_counts().plot()
```

```python
bb["Close"].apply(rank_performance).value_counts().plot(kind = "line")
```

```python
bb["Close"].apply(rank_performance).value_counts().plot(kind = "bar")
```

```python
bb["Close"].apply(rank_performance).value_counts().plot(kind = "barh")
```

### Creating Pie Charts to Represent Proportions

```python
import yfinance as yf
yf.pdr_override()
# y_symbols = ['SCHAND.NS', 'TATAPOWER.NS', 'ITC.NS']
y_symbols = ['BB']
from datetime import datetime
startdate = datetime(2007,1,1)
enddate = datetime(2020,12,31)
bb = pdr.get_data_yahoo(y_symbols, start=startdate, end=enddate)
bb
```

```python
bb["Close"].mean()
```

```python
def rank_performance2(stock_price):
    if stock_price >= 29.459240453332562:
        return "Above Avarage"
    else:
        return "Below Avarage"
```

```python
bb["Close"].apply(rank_performance2)
```

```python
bb["Close"].apply(rank_performance2).value_counts()
```

```python
plt.style.use("dark_background")
```

```python
bb["Close"].apply(rank_performance2).value_counts().plot()
```

```python
bb["Close"].apply(rank_performance2).value_counts().plot(kind = "pie")
```

```python
bb["Close"].apply(rank_performance2).value_counts().plot(kind = "pie", legend = True)
```
